## Imports

**Specify file locations in cell below...**

```python
csvFromEmissions = "C:/Users/david/Documents/00_VSP/01_Mosaik-2/positionEmissionOutput.csv"
csvFromNetCdf = "C:/Users/david/Documents/00_VSP/01_Mosaik-2/positionNetCDFOutput.csv"
```

**Import dataframes...**

```python
df_from_emissions = pd.read_csv(csvFromEmissions, sep=";")
df_from_emissions.head()
```

```python
df_from_netcdf = pd.read_csv(csvFromNetCdf, sep=";")
df_from_netcdf.head()
```

## Some adjustments to make comparable

```python
df_from_emissions.shape[0]
```

```python
df_from_netcdf.shape[0]
```

**Remove empty rows from df_from_netcdf**

```python
df_from_netcdf = df_from_netcdf[df_from_netcdf.CO!=0]
```

```python
df_from_netcdf.shape[0]
```

**Adjust column vehicle_id**

```python
df_from_emissions["vehicle_id"] = df_from_emissions["vehicle_id"].apply(lambda x: int(x.replace("_car","")))
```

```python
df_from_emissions.head(1)
```

**Adjust column time**

```python
df_from_emissions["time"] = df_from_emissions["time"].apply(lambda x: int(x))
```

```python
df_from_emissions.head(1)
```

## Compare in Numbers

**No of missing events in netcdf**

```python
df_from_emissions.shape[0]-df_from_netcdf.shape[0]
```

```python
pd.merge(df_from_emissions,df_from_netcdf, on=['time','vehicle_id'], how="outer")
```
